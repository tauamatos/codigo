```python
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Polygon
```

```python
# Coordenadas fornecidas
coordinates = [
    (-38.51329011531235, -12.98677476861806),
    (-38.51300234252825, -12.987143857805224),
    (-38.514369878151854, -12.987592036795318),
    (-38.51440923169498, -12.987465012885579),
    (-38.51419278720779, -12.987314022115932),
    (-38.51410424173576, -12.987393112530523)
]
```

```python
# Criar o polígono
polygon = Polygon(coordinates)

# Criar um GeoDataFrame
gdf = gpd.GeoDataFrame([{'geometry': polygon}], crs="EPSG:4674")  # SIRGAS 2000
gdf = gdf.rename(columns={0: 'geometry'})
```

```python
# Plotar o polígono
fig, ax = plt.subplots()
gdf.boundary.plot(ax=ax)
ax.set_title('Polígono com Coordenadas no Sistema SIRGAS 2000')
plt.show()
```

**BAIXAR SHAPEFILE**

```python
# Salvar o GeoDataFrame como Shapefile
shapefile_path = '/content/polygon.shp'
gdf.to_file(shapefile_path)

# Compactar os arquivos do shapefile em um zip
shapefile_zip_path = '/content/polygon_shapefile.zip'
with zipfile.ZipFile(shapefile_zip_path, 'w') as shapefile_zip:
    shapefile_zip.write('/content/polygon.shp')
    shapefile_zip.write('/content/polygon.shx')
    shapefile_zip.write('/content/polygon.dbf')
    shapefile_zip.write('/content/polygon.prj')

# Baixar o arquivo zip contendo o Shapefile
files.download(shapefile_zip_path)
```
